```python
#1. NUMPY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gerar 7 valores aleatórios entre 20 e 35 (temperaturas)
temperaturas = np.random.randint(20, 36, size=7)

# Dias da semana
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"]

# 2. PANDAS
# Criar DataFrame
df = pd.DataFrame({"Dia": dias, "Temperatura (°C)": temperaturas})

# Calcular média da semana
media = df["Temperatura (°C)"].mean()
print("Temperaturas registradas:\n", df)
print(f"\nMédia da semana: {media:.2f}°C")

# 3. MATPLOTLIB
plt.plot(dias, temperaturas, marker="o", linestyle="-", color="b")
plt.title("Variação da Temperatura ao Longo da Semana")
plt.xlabel("Dias da Semana")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()
```
